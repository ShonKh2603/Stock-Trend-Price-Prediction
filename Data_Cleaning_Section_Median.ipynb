```python
import pandas as pd
import numpy as np
```

```python
#Loading the whole dataframe that we crawled in the first section of the project
df = pd.read_csv('Complete_df.csv')
```

```python
#Renaming the col names of the percents to remember
df = df.rename(columns={'Held_By_Institusion':'Held_By_Institusion(%)', 'Perf_Week':'Perf_Week(%)','Perf_Month':'Perf_Month(%)', 'Perf_Quarter':'Perf_Quarter(%)','Perf_Half_Year':'Perf_Half_Year(%)','Perf_Year':'Perf_Year(%)'})
```

```python
#converting everything from object to str in order to be able to split the k M and %
df['Vol_Avg_10_Day'] = df['Vol_Avg_10_Day'].astype(str)
df['Vol_Avg_3_Months'] = df['Vol_Avg_3_Months'].astype(str)
```

```python
#We want to replace every "" sign and % sign from the data frame to revert to float after!
df.replace({'%':'' , ',':''} , regex=True, inplace=True)
```

```python
#We want to take care on k(Thousand) and M(millions) (Vol_Avg_10_days)
#We have that in column 6
for index in range(5399):
    if("k" in df.iloc[index,6]):
        df.iloc[index,6] = float(df.iloc[index,6].split('k')[0])*1000
    elif("M" in df.iloc[index,6]):
        df.iloc[index,6] = float(df.iloc[index,6].split('M')[0])*1000000
```

```python
#We want to take care on k(Thousand) and M(millions) (Vol_Avg_3_month)
#We have that in column 7
for index in range(5399):
    if("k" in df.iloc[index,7]):
        df.iloc[index,7] = float(df.iloc[index,7].split('k')[0])*1000 #From object to Float
    elif("M" in df.iloc[index,7]):
        df.iloc[index,7] = float(df.iloc[index,7].split('M')[0])*1000000 #From object to Float
```

```python
#Every sign of '-' we want to put nan instead because it is the same (the item is missed)!
#After we put nan instead of '-' we can handle with all nan at the same time and not with '-' and the wit nan!
df[df == '-'] = np.nan 
```

```python
#All the cols with a % sign
cols = ['Held_By_Institusion(%)' ,'Perf_Week(%)', 'Perf_Month(%)', 'Perf_Quarter(%)','Perf_Half_Year(%)', 'Perf_Year(%)']
```

```python
#Reverting everything to float for the machine learning section only numbers and not strings or objects!
df['Avg_50'] = df['Avg_50'].astype(float)
df['Avg_200'] = df['Avg_200'].astype(float)
df['Held_By_Institusion(%)'] = df['Held_By_Institusion(%)'].astype(float)
df['Perf_Week(%)'] = df['Perf_Week(%)'].astype(float)
df['Perf_Month(%)'] = df['Perf_Month(%)'].astype(float)
df['Perf_Quarter(%)'] = df['Perf_Quarter(%)'].astype(float)
df['Perf_Half_Year(%)'] = df['Perf_Half_Year(%)'].astype(float)
df['Perf_Year(%)'] = df['Perf_Year(%)'].astype(float)
df['Week_52_High'] = df['Week_52_High'].astype(float)
df['Week_52_Low'] = df['Week_52_Low'].astype(float)
df['Vol_Avg_10_Day'] = df['Vol_Avg_10_Day'].astype(float)
df['Vol_Avg_3_Months'] = df['Vol_Avg_3_Months'].astype(float)
df['Analyst_Mean_Target_Price'] = df['Analyst_Mean_Target_Price'].astype(float)
df['Current_Stock_Price'] = df['Current_Stock_Price'].astype(float)
df['Last_Q_Net'] = df['Last_Q_Net'].astype(float)
df['Prev_Q_Net'] = df['Prev_Q_Net'].astype(float)
```

```python
Saving_Symbols = df['Symbols']
#Dropping the unnamed column that we didnt build so we removed it from the dataframe
df.drop('Unnamed: 0', axis = 1, inplace = True)
#Dropping the Symbols column in order to not get dusturbed while we fill ne Nan's
df.drop('Symbols' , axis=1, inplace = True)
#Dropping all the rows that the columns are Nan!!
df.dropna(how='all', inplace = True)
#Filling the missing sectors as unknown
df['Sector'].fillna('unknown', inplace = True)
#We have ' ' as well so we replacing everything to unknown
df['Sector'].replace('', 'unknown', inplace = True)
#Filling the numbers (After the dot 2 numbers) Before this we had for example 53.0700000000!
#Our teacher told us to make the median instead mean
df.fillna(df.median().round(2), inplace = True)
```

```python
#Dividing all the % cols by 100!
df[cols] = df[cols].div(100)
```
